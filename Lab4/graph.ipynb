```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "u2_output.txt"

data = pd.read_csv(file_path, delimiter = ',')

x   = data.iloc[:, 0]  
u_x = data.iloc[:, 1]  

plt.figure(figsize = (8, 6))
plt.plot(x, u_x, label = "u_i", color = "blue", marker = "o", markersize = 0.5)

plt.xlabel("x_i")
plt.ylabel("u_i")
plt.legend()
plt.grid(True)

plt.show()
```
